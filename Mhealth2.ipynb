```python
import pandas as pd
#Read in the dataset
df1 = pd.read_csv('C:/datasets/mHealth.csv', encoding = 'latin1')
#Latin 1 encoding is to interpret special character, to read in the file without producing error

df1
```

```python
df2 = df1.drop("q3t6", axis=1)#Delete "q3t6" column from the dataframe
df2.head()
```

```python
df2.info()
```

```python
df2 = df2.drop_duplicates()
df2
```

```python
df= df2.dropna()# To drop all rows that are completely empty
```

```python
# List for character to remove
chars_to_remove = ['/']
# List for column name to clean
cols_to_clean = ['DAY']

# Loop for each column in cols_to_clean
for col in cols_to_clean:
    # Loop for each char in chars_to_remove
    for char in chars_to_remove:
        # Replace the character with an empty string
        df[col] = df[col].apply(lambda x: x.replace(char,''))
        
# Print a summary of the dataframe
print(df.head())
```

```python
df.info()
```

```python
#Exploring JOB
import matplotlib.pyplot as plt
series = df["JOB"].value_counts()
series
```

```python
series.plot(kind='bar')
plt.title("Number of People that use the app by each Occupation")
plt.xlabel("Occupation")
plt.ylabel("FREQUENCY")

plt.show()
```
